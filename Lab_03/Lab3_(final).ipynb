````python
def exploratory_data_analysis(dataframe):
    # (1) Dimensiones
    filas, columnas = dataframe.shape
    print(f"1.- El DataFrame tiene {filas} filas y {columnas} columnas.")

    # (2) Columnas del DataFrame
    print(f"\n2.- El DataFrame est√° compuesto por las siguientes columnas: {dataframe.columns.tolist()}")

    # (3) Ejemplos de filas del DataFrame
    print("\n3.- Ejemplos de filas del DataFrame:")
    print("\nPrimeras 5 filas:")
    display(dataframe.head())
    print("\n√öltimas 5 filas:")
    display(dataframe.tail())
    print("\nMuestreo aleatorio de 5 filas:")
    display(dataframe.sample(5))

    # (4) Descripci√≥n num√©rica del DataFrame
    print("\n4.- Descripci√≥n num√©rica del DataFrame:")
    display(dataframe.describe(include='number'))

    # (5) Cantidad de valores nulos por columna
    print("\n5.- Cantidad de valores nulos por columna:")
    display(dataframe.isna().sum())

    # (6) Cantidad de valores √∫nicos por columna
    print("\n6.- Cantidad de valores √∫nicos por columna:")
    display(dataframe.nunique())

    # (7) Total de filas que contengan nombres duplicados
    total_duplicados = dataframe.duplicated(subset=['names']).sum()
    print(f"\n7.- Presenta un total de {total_duplicados} filas con nombres duplicados.")

    # (8) Las filas con nombres duplicados
    print("\n8.- Las filas con nombres duplicados son:")
    display(dataframe[dataframe.duplicated(subset=['names'], keep='first')])

    return
````

> Utilice la funci√≥n creada en el punto 1.1 para verificar sus resultados.

````python
def clean_data(df):
    # Eliminar filas duplicadas
    df = df.drop_duplicates(subset='names')

    # Eliminar filas con valores nulos
    df = df.dropna()


    # Filtrar puntajes que van de 0 a 100
    df = df[(df['math score'] >= 0) & (df['math score'] <= 100)]
    df = df[(df['reading score'] >= 0) & (df['reading score'] <= 100)]
    df = df[(df['writing score'] >= 0) & (df['writing score'] <= 100)]

    # Eliminar filas con mismos nombres pero puntajes distintos
    df = df.groupby('names').filter(lambda x: x[['math score', 'reading score', 'writing score']].nunique().max() == 1)

    return df

````

````python
df_grades = clean_data(df_grades)
df_grades
````

**Justificaci√≥n:**

```
1. Se elimina los duplicados por nombres, para que un estudiante no rinda m√°s de dos veces una prueba.
2. Se eliminan los NA.
3. Se filtra por m√°ximo y m√≠nimo de notas (posible, 0 a 100).
```

### 3.1 Transformaci√≥n de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci√≥n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci√≥n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci√≥n entregada, cree una funci√≥n que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):

  for x in columna:
    dataframe_in[x] = dataframe_in[x].apply(lambda x: round(x  * 0.06 + 1, 2))
  return dataframe_in
````

### 3.2 Obtenci√≥n de promedio de alumnos [0.3 Puntos]

Con todas las notas adjuntadas, la prestigiosa Universidad De la Cacha√±a necesita obtener el promedio general de cada uno de los alumnos. Para esto se le solicita que calcule el promedio con cada una de las asignaturas y el promedio se anexe como una nueva columna al final del `DataFrame` original. Cabe se√±alar que las notas deben ser redondeadas al segundo decimal.

**To-Do:**
- [ ] Crear funci√≥n que nos permite obtener el promedio de cada uno de los alumnos y anexar esta informaci√≥n en el `DataFrame` original.


**Respuesta:**

````python
def alumnos_promedio(dataframe_in):
    # Calcular el promedio de cada alumno
    promedio_alumnos = dataframe_in[['math score', 'reading score', 'writing score']].mean(axis=1)

    # Redondear el promedio al segundo decimal
    promedio_alumnos = promedio_alumnos.round(2)

    # Anexar el promedio como una nueva columna al DataFrame original
    dataframe_in['GPA'] = promedio_alumnos

    return dataframe_in
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Para esto se considerar√° que un alumno ha reprobado el a√±o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar√° destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a√±o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    df_reprobados = dataframe_in[dataframe_in['GPA']<4]

    df_R = dataframe_in.copy()
    df_R.loc[df_R['GPA']<4, 'GPA'] = 'R'

    df_destacados = dataframe_in.loc[(dataframe_in['math score']>5.8) & (dataframe_in['reading score']>5.8) & (dataframe_in['writing score']>5.8)]
    df_destacados = df_destacados.sort_values(by=['GPA'], ascending=False)

    return df_reprobados, df_R, df_destacados
````

````python
# Calcular el promedio de las notas y guardarlo en una variable GPA
df_sg = alumnos_promedio(df_sg)

#Hacer una simplificaci√≥n a trav√©s de un mapeo de la variable
parental_level_dic = {
    "some high school" : "school",
    "some college" : "school",
    "high school" : "school",
    "bachelor's degree" : "college",
    "associate's degree" : "college",
    "master's degree" : "postgraduate"
}

df_sg['simple parental level of education'] = df_sg['parental level of education'].map(parental_level_dic)

#Agregar seg√∫n 2 niveles: race/ethnicity y simple parental level of education para obtener el promedio de las notas.
tabla_resumen = df_sg.groupby(['race/ethnicity', 'simple parental level of education'])[['math score','reading score', 'writing score', 'GPA']].mean().reset_index()

# Agregar seg√∫n 2 niveles: race/ethnicity y simple parental level of education para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
df_count = df_sg.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')
tabla_resumen['count'] = df_count['count']

#Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.
N = len(df_sg)
tabla_resumen['porcentage'] = (100 * tabla_resumen['count'] / N).round(2).astype(str) + '%'
````

````python
fig = px.box(
    tabla_resumen,
    x="simple parental level of education",
    y="GPA",
    color="race/ethnicity",
    hover_data=["GPA"],
)
fig.show()
````
